```python
#to be in line with the datasets below, waste generated needs to be in million tonnes
waste_3to17 = pd.read_csv('./datasets/2003_2017_waste.csv')
waste_3to17[['waste_disposed_of_tonne', 'total_waste_recycled_tonne', 'total_waste_generated_tonne']] = waste_3to17[['waste_disposed_of_tonne', 'total_waste_recycled_tonne', 'total_waste_generated_tonne']].apply(lambda x: round((x/1000000), 5))

#changing the name of the columns to correspond with the values
waste_3to17 = waste_3to17.rename({'waste_disposed_of_tonne': 'total_disposed_mil_tonne', 'total_waste_recycled_tonne': 'total_recycled_mil_tonne', 'total_waste_generated_tonne': 'total_generated_mil_tonne'},axis = 1)

#sort by year (as 2017 is after 2013)
waste_3to17 = waste_3to17.sort_values('year', ignore_index=True)

waste_3to17.head(5)
```

```python
#to be in line with the datasets below, waste generated needs to be in million tonnes
# waste_18to20 = pd.read_csv('./datasets/2018_2020_waste.csv')
waste_18to20[['Total Generated (\'000 tonnes)', 'Total Recycled (\'000 tonnes)']] = waste_18to20[['Total Generated (\'000 tonnes)', 'Total Recycled (\'000 tonnes)']].apply(lambda x: round((x/1000), 5))

#changing the name of the columns to correspond with the values
waste_18to20 = waste_18to20.rename({'Waste Type': 'waste_type', 'Total Generated (\'000 tonnes)': 'total_generated_mil_tonne', 'Total Recycled (\'000 tonnes)': 'total_recycled_mil_tonne', 'Year': 'year'},axis = 1)

#adding new columns for total disposed and recycling rate
waste_18to20 = waste_18to20.assign(total_disposed_mil_tonne=lambda x: x.total_generated_mil_tonne - x.total_recycled_mil_tonne)
waste_18to20 = waste_18to20.assign(recycling_rate=lambda x: round((x.total_recycled_mil_tonne / x.total_generated_mil_tonne),2))

#rearranging columns
waste_18to20 = waste_18to20[['waste_type', 'total_disposed_mil_tonne', 'total_recycled_mil_tonne', 'total_generated_mil_tonne', 'recycling_rate', 'year']]

#sort by year
waste_18to20 = waste_18to20.sort_values('year', ignore_index=True)

waste_18to20.head(5)
```

```python
#join data from the two datasets
waste_Total = pd.concat([waste_3to17, waste_18to20], ignore_index=True)
waste_Total
```

```python
#standardizing all values for waste type
waste_Total = waste_Total.replace(['Horticultural waste', 'Horticultural'], 'Horticultural Waste')
waste_Total = waste_Total.replace('Plastic', 'Plastics')
waste_Total = waste_Total.replace(['C&D', 'Construction debris', 'Construction& Demolition', 'Construction & Demolition'], 'Construction Debris')
waste_Total = waste_Total.replace('Wood/Timber', 'Wood')
waste_Total = waste_Total.replace(['Ferrous Metal', 'Ferrous metal'], 'Ferrous Metals')
waste_Total = waste_Total.replace(['Food', 'Food waste'], 'Food Waste')
waste_Total = waste_Total.replace(['Non-ferrous metal', 'Non-ferrous metals', 'Non-Ferrous Metal'], 'Non-ferrous Metals')
waste_Total = waste_Total.replace(['Sludge', 'Ash and sludge', 'Ash & sludge'], 'Ash & Sludge')
waste_Total = waste_Total.replace('Scrap tyres', 'Scrap Tyres')
waste_Total = waste_Total.replace(['Others (stones, ceramics & rubber etc)', 'Others (stones, ceramics & rubber etc.)', 'Others (stones, ceramic, rubber, etc.)', 'Others (stones, ceramic, rubber, ect)', 'Others (stones, ceramics, etc.)'], 'Others')
waste_Total = waste_Total.replace('Overall', 'Total')
waste_Total = waste_Total.replace('Used slag', 'Used Slag')
```

```python
#formatting table to a readable format
co2emission = pd.read_json('./datasets/singapore-carbon-co2-emissions.json')

#dropping rows full of words
co2emission.drop(index=co2emission.index[:22], inplace=True)
#dropping columns that are empty
co2emission.drop(columns=co2emission.columns[-18:], inplace=True, axis=1)

#splitting values by comma to get 3 different values
co2emission[['year', 'co2_emission', 'metric_tons_per_capita']] = co2emission.Macrotrends.str.split(",", expand = True)
co2emission[['year', 'Month', 'Day']] = co2emission.year.str.split("-", expand = True)

#removing columns that are not in use
co2emission.drop(columns=['Macrotrends','metric_tons_per_capita','Month','Day'], inplace=True, axis=1)

#changing kiloton to million tonnes
co2emission = co2emission.astype({'co2_emission': 'float64'})
co2emission['co2_emission'] = co2emission['co2_emission'].apply(lambda x: round((x/1000), 2))
#adding data found from nea
co2emission.loc[len(co2emission.index)] = ['2020', 46.75] 

co2emission = co2emission.reset_index(drop=True)
co2emission = co2emission.set_index('year')
co2emission
```

```python
#text based analysis on different types of wastes
waste_composition = waste_Total[waste_Total.waste_type != 'Total']

# Group the data by waste_type
grouped_df = waste_composition.groupby("waste_type")
mean = grouped_df[["total_disposed_mil_tonne", "total_recycled_mil_tonne", "total_generated_mil_tonne", "recycling_rate"]].mean()

mean = pd.DataFrame(mean)

def highest(data):
    mean.sort_values(data, ascending=False, inplace=True)
    most_waste_type = mean.index[0]
    most_waste_type_val = round(mean.loc[most_waste_type][data], 5)
    print(f'Waste type with the most {data}: {most_waste_type} ({most_waste_type_val})')

def lowest(data):
    mean.sort_values(data, ascending=True, inplace=True)
    least_waste_type = mean.index[0]
    least_waste_type_val = round(mean.loc[least_waste_type][data], 5)
    print(f'Waste type with the least {data}: {least_waste_type} ({least_waste_type_val})')

print('Text Based Analysis on waste composition (average from 2003 to 2020)')
print('-'*70)

highest("total_generated_mil_tonne")
highest("total_disposed_mil_tonne")
highest("total_recycled_mil_tonne")
highest("recycling_rate")

print('')

lowest("total_generated_mil_tonne")
lowest("total_disposed_mil_tonne")
lowest("total_recycled_mil_tonne")
lowest("recycling_rate")
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'total_disposed': mean['total_disposed_mil_tonne'],
    'total_recycled': mean['total_recycled_mil_tonne']
})

# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_

data['Cluster'] = labels
text_data = data.groupby("Cluster")

for cluster, group in text_data:
    print("Cluster: ", cluster)
    print("Waste Types: ", group.index)
    print("Mean Total Disposed Waste: ", group['total_disposed'].mean())
    print("Mean Total Recycled Waste: ", group['total_recycled'].mean())
    print("\n")

# Plot the data with different colors for each cluster
plt.scatter(data['total_disposed'], data['total_recycled'], c=labels)
plt.xlabel('total_disposed (mil_tonnes)')
plt.ylabel('total_recycled (mil_tonnes)')
plt.title('Cluster Analysis on amount of disposed and recycled waste by waste type')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
waste_disposed = np.array(waste_by_year['total_disposed_mil_tonne'])
co2_emission = np.array(co2emission['co2_emission'])

model_df = pd.DataFrame({
    "co2_emission": co2_emission,
    "waste_disposed": waste_disposed
})
```

```python
X_train, X_test, y_train, y_test = train_test_split(model_df[['waste_disposed']], model_df[['co2_emission']], test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)
result = model.score(X_test, y_test)
print("Accuracy: %.2f%%" % (result*100.0))
```

```python
import statsmodels.api as sm
```

```python
stockModel = sm.OLS.from_formula("co2_emission ~ waste_disposed", data=model_df).fit()
stockPredicts = stockModel.predict()

# The linear model
print(stockModel.summary())
```
